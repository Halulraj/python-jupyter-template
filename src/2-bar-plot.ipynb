### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Line plot
x = np.arange(0, 11)
y = x**2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')
plt.title('Line Plot of y = x^2')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()
plt.show()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.title('Bar Plot of Categories and Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
